```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
sheet_names= ['loan_information', 'Employment','Personal_information', 'Other_information' ]

data = pd.read_excel('inputs/Credit_Risk_Dataset.xlsx', sheet_names)
```

```python
loan_information = data['loan_information']
employment = data['Employment']
personal_information = data['Personal_information']
other_information = data['Other_information']
```

```python
loan_information.head()
```

```python
data_nan = (loan_information.isnull().sum() / len(loan_information)) * 100
data_nan = data_nan.drop(data_nan[data_nan == 0].index).sort_values(ascending=False)[:10]
missing_data = pd.DataFrame({'Missing Ratio' :data_nan})
missing_data.head(5)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
plt.xticks(rotation='vertical')
sns.barplot(x=data_nan.index, y=data_nan)
plt.xlabel('Features ', fontsize=10)
plt.ylabel('Percent of missing features values', fontsize=10)
plt.title('Percent missing data by feature in loan information dataset', fontsize=15)
```

```python
data_nan = (employment.isnull().sum() / len(employment)) * 100
data_nan = data_nan.drop(data_nan[data_nan == 0].index).sort_values(ascending=False)[:10]
missing_data = pd.DataFrame({'Missing Ratio' :data_nan})
missing_data.head(5)
```
